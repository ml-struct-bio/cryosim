```python
import pickle
import os
import re
import numpy as np
```

```python
def natural_sort_key(s):
    # Convert the string to a list of text and numbers
    parts = re.split('([0-9]+)', s)
    
    # Convert numeric parts to integers for proper numeric comparison
    parts[1::2] = map(int, parts[1::2])
    
    return parts
```

```python
def stack_pickles(folder_path, sorted_pkl):
    stacked_data1 = None
    stacked_data2 = None
    
    # Loop through each file in the folder
    for filename in sorted_pkl:
        print(filename)
        filepath = os.path.join(folder_path, filename)
        print('filepath:',filepath)
        # Check if the file is a Pickle file
        if filename.endswith('.pkl'):
            with open(filepath, 'rb') as file:
                data = pickle.load(file)
                print('data[0][0]:',data[0][0].shape)

                # Stack the data
                if stacked_data1 is None:
                    stacked_data1 = data[0]
                    stacked_data2 = data[1]
                else:
                    stacked_data1 = np.vstack((stacked_data1, data[0]))
                    stacked_data2 = np.vstack((stacked_data2, data[1]))
                    print('stacked_data1:',stacked_data1.shape)
#                     print('stacked_data2:',stacked_data2.shape)

    return (stacked_data1, stacked_data2)
```
